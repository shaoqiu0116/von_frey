```python
import pandas as pd
import pandas as pd

df1 = pd.read_csv('forcelookup_table.csv', index_col='No.')
print(df1.head(10))

df2 = pd.read_csv('df2.csv', index_col='No.')
print(df2.head(10))

df3 = df2.reset_index().merge(
    df1, on='Response', how="left").set_index(
        'No.'
    )  # Reset_index keep the origninal index, and then set to the new index
print(df3.head(10))
df3.to_csv('Result.csv')
```

```python
import pandas as pd

df1 = pd.read_csv('forcelookup_table.csv', index_col='No.')
print(df1.head())

df2 = pd.read_csv('df2.csv', index_col='No.')
print(df2.head())

df2['Calculated_Value'] = pd.Series()

df2['Calculated_Value'] = df2.apply(
    lambda x: df1.loc[x['Response'] == df1['Response'], 'Force'].reset_index(
        drop=True),
    axis=1)
print(df2)
df2.to_csv('df3.csv')
```

```python
import pandas as pd

df = pd.read_csv('df1.csv')
print(df.head())
df = df.drop_duplicates(subset='Response')
df = df.sort_values(by='Force')
df = df.reset_index(drop=True)
print(df.head())
df.to_csv('forcelookup_table.csv')
```

```python
import pandas as pd

df1 = pd.DataFrame({
    'Response': ['XXX', 'XXO', 'OOX', 'XOX'],
    'Value': [2.406, 2.534, 2.668, 3.124]
})
print(df1)

df2 = pd.DataFrame({'Response': ['XXX', 'XOX', 'XOX', 'XXO', 'XOX']})
print(df2)

df2['Value'] = pd.Series()

df2['Value'] = df2.apply(
    lambda x: df1.loc[x['Response'] == df1['Response'], 'Value'].reset_index(
        drop=True),
    axis=1)
print(df2)
df1.to_csv('df4.csv')
```

```python
df1 = pd.DataFrame({
    'Year': [1910, 1911, 1912],
    'CA': [2.406, 2.534, 2.668],
    'HI': [0.804, 0.821, 0.832]
})
print(df1)

df2 = pd.DataFrame({
    'State': ['CA', 'CA', 'CA', 'HI', 'HI'],
    'Year': [1910, 1910, 1911, 1911, 1911]
})
print(df2)

df2['Population'] = pd.Series()

df2['Population'] = df2.apply(lambda x: df1.loc[x['Year'] == df1['Year'], x[
    'State']].reset_index(drop=True),
                              axis=1)

print(df2)
```
